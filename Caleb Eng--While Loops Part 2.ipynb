### 1. Continue Statement

In Python, the <code> continue </code> statement is used inside loops to skip the rest of the current iteration and move directly to the next one. When Python encounters <code> continue </code>, it immediately stops executing any remaining code in that loop cycle and jumps back to check the loop condition or move to the next item. This is useful when you want to ignore certain cases without stopping the entire loop. For example, if you're looping through numbers and want to skip even numbers, you can use continue to bypass them while still continuing the loop for the remaining values.

Try out this example below:

```python
# Code Cell 1

num = 0
while num < 10:
    num += 1
    if num % 2 == 0:
        continue  # skip even numbers
    print(num)
```

### 2. Break Statement

In Python, the <code> break </code> statement is used to immediately exit a loop, regardless of whether the loop condition is still true. When Python encounters a  <code> break </code>, it stops the loop entirely and continues running the code that comes after the loop. This is useful when you are searching for something or waiting for a specific condition, and once it’s met, there’s no need to keep looping. Unlike continue, which skips just one iteration, break ends the whole loop.

Try out this example:



```python
# Code Cell 2

num = 1
while True:
    if num == 5:
        break  # stop the loop when num is 5
    print(num)
    num += 1

print("All done!")
```

### 3. While + Else

In Python, a <code> while </code>  loop can have an <code>  else </code> block that runs after the loop finishes only if the loop was not stopped by a break statement. This means the else acts like a “completed normally” signal—if the loop condition becomes false and the loop ends naturally, the else code will execute. However, if the loop is exited early using break, the else block is skipped. This is especially useful in situations like searching, where you want to know whether something was found or not.

Try out these examples:



```python
# Code Cell 3

num = 1

while num <= 5:
    print(num)
    num += 1
else:
    print("Loop finished normally!")
```

```python
# Code Cell 4

target = 7
num = 1

while num <= 10:
    print(num)
    if num == target:
        print("Found the number!")
        break
    num += 1
else:
    print("Number not found.")
```
