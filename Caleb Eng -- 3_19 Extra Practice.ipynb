### Why or when Do We Nest if/else Statements?

Sometimes a decision depends on another decision. In these cases, we nest an if/else statement inside another one to handle more complex logic step by step. Think of it like a flowchart: first you answer one question, and depending on that answer, you move on to a second question. Nesting helps us organize these multi-step decisions clearly in code.

For example:

```python
# Code Cell 1
age = 16
has_permit = True

if age >= 16:
    if has_permit:
        print("You can drive!")
    else:
        print("You need a permit to drive.")
else:
    print("You are too young to drive.")
```

In this example, the program first checks the age. Only if that condition is true does it check the second condition (permit).

### What is elif?

The elif statement (short for “else if”) is used when you want to check multiple conditions in order, without deeply nesting multiple if statements. It makes code cleaner and easier to read when you have several possible cases.

Let's see how this works:

```python
# Code Cell 2:

grade = 85

if grade >= 90:
    print("A")
elif grade >= 80:
    print("B")
elif grade >= 70:
    print("C")
else:
    print("Needs improvement")
```
