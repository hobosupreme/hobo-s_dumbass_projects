## The <code>for</code> Loop

In other coding languages like C++ and Java, <code>for</code>-loops are usually introduced first. In those languages, the syntax of the <code>for</code>-loop looks like this:
<pre>
for (int i = 0; i < N; i += 1)
{
    do something;
    do something;
}
</pre>
This is a more compact way of handling the loop counter variable <code>i</code> - it is defined, initialized, incremented, and compared with the number of times we want to iterate the loop, all in an expression following the keyword <code>for</code>. However, in Python, this construct does not exist as such. Instead, the loop counter is *generated* by a so-called <code>range</code> object. This saves us the trouble of initializing and incrementing a loop counter variable inside the <code>for</code>-block.

In Python, <code>range</code> generates a sequence of integers, and can take a number of forms:

* <code>range(<font color="blue"><b>stop</b></font>)</code>: generates integers ```0```, ```1```, ... , <code><font color="blue"><b>stop-1</b></font></code> (be careful here - <b>this is NOT <code><font color="blue"><b>stop</b></font></code></b>!)
* <code>range(<font color="red"><b>start</b></font>, <font color="blue"><b>stop</b></font>)</code>: generates integers <code><font color="red"><b>start</b></font></code>, <code><font color="red"><b>start</b></font>+1</code>, ... , <code><font color="blue"><b>stop-1</b></font></code> (again, be careful - <b>this is NOT <code><font color="blue"><b>stop</b></font></code></b>!)
* <code>range(<font color="red"><b>start</b></font>, <font color="blue"><b>stop</b></font>, <font color="green"><b>step</b></font>)</code>: generates integers <code><font color="red"><b>start</b></font></code>, <code><font color="red"><b>start</b></font>+<font color="green"><b>step</b></font></code>, ... , <code><font color="red"><b>start</b></font>+n*<font color="green"><b>step</b></font></code> (```n``` is an integer such that <code><font color="red"><b>start</b></font> + n*<font color="green"><b>step</b></font> < <font color="blue"><b>stop</b></font></code>)

In all of these, <code><font color="red"><b>start</b></font></code>, <code><font color="blue"><b>stop</b></font></code>, and <code><font color="green"><b>step</b></font></code> <b>MUST</b> all be integers. If ```start >= stop``` and ```step > 0```, no integers will be generated by ```range()```. Also, ```step``` must be nonzero, or an error will result.

This is used in a <code>for</code>-loop, which looks similar to <code>if</code>- and <code>while</code>-blocks, and takes the general form:
<pre>
for i in range(<font color="red"><b>start</b></font>, <font color="blue"><b>stop</b></font>, <font color="green"><b>step</b></font>):
    do something
    do something
</pre>

If you find this confusing, see <a href="https://www.w3schools.com/python/ref_func_range.asp">Python <code>range()</code> Function</a>. <br>

To see how this works, let's run the following code cell:

````python
# Code Cell 1

for i in range(1, 6):
    print(i)
````

See that the code above has a range from 1 to 6. Each time the program loops, i is updated to the next item in the range, and is printed.

Here is another example but this time, the range is a string.

````python
# Code Cell 2

for letter in "hello":
    print(letter)
````

In example #2, the for loop works by treating the string "hello" as a sequence of individual characters. In Python, strings are iterable, which means you can go through them one piece at a time—just like a list. When the loop runs, it takes the first character "h" and assigns it to the variable letter, then executes the indented code (in this case, print(letter)). It then moves to the next character "e", then "l", "l", and finally "o", repeating the same process each time. The loop automatically stops once it reaches the end of the string, so each character gets printed on its own line without needing to manually track positions or indexes.

Here is one more example, showing how you could use a for loop to add numbers, each time it loops:

````python
# Code Cell 3

total = 0

for i in range(1, 6):
    total += i

print(total)
````

<code> break </code>and <code>continue</code> are both used to control how a loop runs, but they do very different things.

Like with while loops, <code> break </code> completely stops the loop early, even if there are still items left to go through. <code> continue </code> skips the current iteration and moves on to the next one.

Here’s how they work in practice:

````python
# Code Cell 4

for i in range(1, 10):
    if i == 5: #this will stop the loop when it reaches 5, even though the range goes to 10
        break
    print(i)
````

````python
# Code Cell 5

for i in range(1, 10):
    if i == 5: # This will skip the number 5, and then keep going!
        continue
    print(i)
````
